**What is Pytesseract??**

Pytesseract can be used as a standalone script for Tesseract allowing it to print recognized text instead of converting it to a file. Pytesseract can read all image files that are supported by imaging libraries such as Leptonica and Pillow, including JPEG, PNG, GIF, BMP, TIFF, and many others.

**What Algorithm does Pytesseract use??**

In general, Tesseract uses adaptive binarization for character recognition from binary format of images and for images with single character, Tesseract uses Convolutional Neural Network (CNN).

**Features** of Pytesseract



*   Scanned PDF to Searchable PDF
*   Reads 60+ languages
*   Unbeatable OCR accuracy
*   Affordable OCR library
*   Well documented SDK.
*   Effortless deployment



**Drawbacks** of Pytesseract



*   Tesseract is not very good at handling low-quality images or documents with low resolution.
*   Accuracy drops because of noise
*   Large Amounts of texts cant be accurately processed
*   Tesseract is prone to errors if the separation of the foreground and background of the image is not significant.
